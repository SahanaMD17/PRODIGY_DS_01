```python
# To read the csv file(dataset)
data = pd.read_csv('population.csv')
data.head()
```

```python
data_final = data.drop(['Country Code',	'Indicator Name',	'Indicator Code'], axis=1)
chosen_country = 'India'

country_data = data_final[data_final['Country Name'] == chosen_country]

years = country_data.columns[2:].astype(int)
population = country_data.iloc[:, 2:].values.flatten()
```

# **BAR CHART**

```python
colors = plt.cm.viridis(np.linspace(0.3, 0.8, len(years)))
plt.figure(figsize=(10, 6))
bars = plt.bar(years, population, color=colors, edgecolor='black', linewidth=0.5)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.title(f'Population of {chosen_country} Over the Years', fontsize=14, color='darkslategray')
plt.grid(axis='y', linestyle='--', alpha=0.5)

for i, (year, pop) in enumerate(zip(years, population)):
    if i % 5 == 0:  # Annotate every 5 years for simplicity
        plt.text(year, pop + 0.03 * pop, f'{int(pop):,}', ha='center', va='bottom',
                 fontsize=9, color='darkgreen')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

# **HISTOGRAM**

```python
plt.figure(figsize=(10, 6))
plt.hist(population, bins=20, color='indianred', edgecolor='black', alpha=0.85)

plt.title('Population Distribution Histogram', fontsize=14, fontweight='bold')
plt.xlabel('Population', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.ticklabel_format(style='plain', axis='x')

# a simple line for the mean
mean_pop = np.mean(population)
plt.axvline(mean_pop, color='orange', linestyle='--', linewidth=1.5)
plt.text(mean_pop * 1.02, plt.ylim()[1] * 0.9, f'Mean: {int(mean_pop):,}', color='orange', fontsize=10)
plt.show()
```

```python
country1 = data_final[data_final['Country Name'] == 'United States'].iloc[:, 2:].values.flatten()
country2 = data_final[data_final['Country Name'] == 'Indonesia'].iloc[:, 2:].values.flatten()

plt.figure(figsize=(10, 6))
plt.hist(country1, bins=30, color='royalblue', edgecolor='black', alpha=0.7, label='United States', density=True)
plt.hist(country2, bins=30, color='orange', edgecolor='black', alpha=0.7, label='Indonesia', density=True)

plt.title('Population Distribution Comparison')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
